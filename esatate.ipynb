 ### Оценка динамики рынка недвижимости СПб в течение нескольких лет (2014 - 2019 гг) 
  
   Исследование посвящено изучению рынка недвижимости СПб и ближайших пригородов на основе систематизации объявлений о продаже объектов недвижимости, опубликованных в диапазоне 2014 - 2019 гг.
 
 ### Цель и задача исследований
    
   Задачей исследования является выявление основных закономерностей в предоставленных данных, связанных с различными переменными факторами: характеритики объектов недвижимости: кол-во комнат, этажность, близость к объектам природным и инфраструктурным объектам, стоимость объектов и динамика их реализации.
    Целью исследований данных является оптимизация риэторских услуг по размещению рекламных объявлений, поддержке клиента, определению стратегии ценообразования и пролонгированию рекламной кампании, определению вероятных временных рамок сделок и оптимизации сопутствующих фременных и финансоавых вложений.
    
 ### Этапы выполнения работы
  
- провести первичную оценку "сырых" данных, выявить основные проблемные места (неадекватные значения данных, неверные названия столбцов, ошибки в названиях нас пунктов, наличие дубликатов и пропусков в столбцах). Провести коррекцию данных, устранитть дубликаты и по возможности заполнить пропуски.
- проверить соответствие типов данных и их физического смысла, при необходимости скорректировать тип данных. 
- выявить взаимную зависимость основных данных (фактических, топографических и временных), влияющих на стоимость объектов. Создать новые столбцы с информацией об удельной стомости объектов (цена кв м, скорость роста цены с приближением центра города и инфраструктурных объектов и пр.). 
- определить четкие критерии ценообразования объекта на основании его расположения, количества комнат, площади объекта, особенностей планировки, этажности дома, близости культурных и природных объектов, расстояния от аэропорта и центра города.
- выявить динамику ценообразования в зависимости от временных факторов  - года, месяца и дня недели размещения объявлений.
- Сделать итоговый вывод об особенностях ценообразования квартир в СПб в исследуемый промежуток времени, на основании которого риэлтор сможет сформулировать четкую стратегию продажи того или иного объекта недвижимости.

###  Ввыводы по пропущенным значениям в столбцах:

1) Пропуски в 'balcony' скорее всего связаны с отсутствием балконов, поэтому их заменили на 0.

2) Пропуски в 'ceiling_height' можно заменить на медианную высоту потолков от всех объектов - большинство объектов имеют высоту потолков в интервале 2,5 - 3 м, то есть замена будет близка к истине.
3) Значения в 'living_area' хорошо коррелируют со значениями в 'total_area' и 'rooms' (коэф-ты Пирса более 0,8), поэтому для удобства группировки, пропуски в 'living_area' были заменены на средние значения для аналогичных объектов, сгруппированных по числу комнат.    
4) Пропуски в булевом 'is_apartment' говорят о том, что объект вероятно не студия, поэтому заменяем их на False.
5) Некоторые пропуски в 'cityCenters_nearest' можно заменить, если соотнести данные этого столбца с данными 'airports_nearest'. Большинство объектов находятся в пределах 30 км от центра города, то есть диаметр города где-то 50 - 60 км (с пригородами диаметр около 80 км). Аэропорт обычно находится в пригороде, то есть, все объекты должны попасть в интервал 80 км от аэропорта, что и происходит на самом деле по данным соответствующей гистограммы. Поэтому, если в интервале 80 км от аэропорта есть объекты с неуказанным расстоянием от центра (но указанным расстоянием до аэропорта), то для них можно указать медианное расстояние до центра, соответствующее медианному расстоянию до центра для объектов, входящих в 80 километровый интервал расстояний от аэропорта (80 км интервал разбивается на интервалы по 10 км). Имеется только 1 такой объект, для которого расстояние до центра было заменено на
медианное значение этого параметра для объектов, лежащих в пределах 10 - 20 км от аэропорта.
6) Для объектов с неизвестным расстоянием от аэропорта, но известным расстоянием до центра (не далее 10 км от центра), также можно заменить пропуски в 'airports_nearest' медианным значением этого параметра для объектов, находящихся в пределах 10 км от центра (для более далеких от центра объектов ошибка будет слишком большой). Подходящих объектов в 10 км от центра и неизвестным расстоянием до аэропорта - 12 шт, и для них расстояние до аэропорта было заменено медианным значением для аналогичных объектов, которые находятся в 10 км от центра.
7) Обращает внимание, что число объектов, для которых не указано расстояние до аэропорта практически полностью совпадает с количеством объектов, для которых не указано также расстояние до центра города. Поэтому действия в пп.5,6 вряд ли повлияют на общую картину. Тем более, замены в пп.5,6 справедливы, если в городе есть только 1 аэропорт. 
8) Для большинства объектов, указавших наличие парков в 3 км зоне, они находится в пределах 1 км. По идее, можно было бы для объектов, указавших наличие парков, но не указавших расстояние до ближайшего, поставить медианное расстояние (по крайней мере, для объектов, указавших 2 и более парков). Но таких объектов нет, то есть, все, кто указал наличие парков указал и расстояние до ближайшего. Поэтому пропуски остаются не заполнены.
9) Для прудов аналогичная ситуация с парками - все, кто указал наличие прудов, указали и расстояние до ближайшего. Так что пропуски остаются.
10) Обращает внимание практически равное количество пропусков в столбцах 'cityCenters_nearest', 'airports_nearest', 'parks_around3000', 'ponds_around3000'. Это значит, что либо эти объекты попали в общую таблицу из другой исходной формы, в которой отсутствовали соответствующие позиции, либо эти категории были заполнены с мобильных устройств, в которых недоступны соответствующие позиции. В любом случае, форма заполнения нуждается в проверке работы на разных устройствах.
11) Пропущенные значения в 'days_exposition' могут быть заменены медианными значениями, характерными для аналогичной категории объектов.
```python
def rapid(w):
    if w <= 45:
        return 'очень быстро'
    elif w > 45 and w <= 92:
        return 'умеренно быстро'
    elif w > 92 and w <= 206:
        return 'средняя скорость'
    elif w > 206 and w <= 433:
        return 'долго, но приемлемо'
    else:
        return 'аномально долго'
data['expo_value'] = data['days_exposition'].apply(rapid)
```
